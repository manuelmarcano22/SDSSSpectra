```python
datastars = []
with open('tableBeaver1980.csv','r') as datafile:
    for lines in datafile:
        datastars.append(lines)
        
datastars = np.array([ i.strip() for i in datastars ])
datastars = [ i.split(',') for i in datastars ]
```
